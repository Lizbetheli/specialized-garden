```python
# Dependencies
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
import json
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.proflowers.com/blog/types-of-flowers'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
pic_flowers=[]
name_flowers=[]
flower_details=[]
results = soup.find_all('figure', class_="row flower")

result= []
# Loop through returned results
for result in results:
    # Error handling
    try:
        flower_detail = result.find('div', class_='row flower_details').text
        name_flower = result.find('div', class_='col-sm-2').text
        pic_flower = result.img['data-src']

        if (pic_flower and name_flower and flower_detail):
            print('-------------')
            print(pic_flower)
            print(name_flower)
            print(flower_detail)
    except AttributeError as e:
        print(e)
        next
    pic_flowers.append(pic_flower)
    name_flowers.append(name_flower)
    flower_details.append(flower_detail)
    
       
```

```python
data_dict={
    'name_flower': name_flowers,
    'flower_detail': flower_details,
    'pic_flower': pic_flowers   
}
df=pd.DataFrame(data_dict)
df
```

```python
# break details apart to multiple columns 
details_df= df.flower_detail.str.split(pat="\n", expand=True)
details_df.head(10)
```

```python
# drop extra columns, rename
new_details = details_df.drop([0, 1,2, 4,5,6,8,9,10,12,13,14,16,17,18,20,21,22,24,25], axis=1)
new_details.rename(columns = {3:'Sun Needs', 7: 'Soil Needs', 11: 'Zones', 15: 'Height', 19:'Blooms in', 23: 'Features'}, inplace = True) 
new_details
```

```python
# join seperated details into main df
join_df = df.join(new_details)
joined_df= join_df.drop(['flower_detail'], axis=1)
# clean up the names from scrape
clean_names= df.name_flower.str.split(pat="\n",expand= True)
cleaned_names = joined_df.join(clean_names)
#drop extra columns,rename, reorganize
org_names= cleaned_names.drop(['name_flower', 0,1,3], axis=1)
rename_column= org_names.rename(columns={ 2: 'Flower Name', 'pic_flower': 'Picture Link'}, inplace=True)
final_df =org_names[['Flower Name', 'Blooms in', 'Zones', 'Sun Needs', 'Soil Needs', 'Height', 'Features', 'Picture Link']]
final_df
```

```python
final_df.to_csv("flowers_data.csv")
```
