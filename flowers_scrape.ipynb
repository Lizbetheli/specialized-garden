```python
# Dependencies
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
import json
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.proflowers.com/blog/types-of-flowers'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
pic_flowers=[]
name_flowers=[]
flower_details=[]
results = soup.find_all('figure', class_="row flower")

result= []
# Loop through returned results
for result in results:
    # Error handling
    try:
        flower_detail = result.find('div', class_='row flower_details').text
        name_flower = result.find('div', class_='col-sm-2').text
        pic_flower = result.img['data-src']

        if (pic_flower and name_flower and flower_detail):
            print('-------------')
            print(pic_flower)
            print(name_flower)
            print(flower_detail)
    except AttributeError as e:
        print(e)
        next
    pic_flowers.append(pic_flower)
    name_flowers.append(name_flower)
    flower_details.append(flower_detail)
    
       
```

```python
data_dict={
    'name_flower': name_flowers,
    'flower_detail': flower_details,
    'pic_flower': pic_flowers   
}
df=pd.DataFrame(data_dict)
```
