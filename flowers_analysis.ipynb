```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
# File to Load, read, store
data_to_load = "flowers_data.csv"
data = pd.read_csv(data_to_load)
data
```

```python
plot = data["Blooms in"].value_counts()
plot.plot(kind = "bar")
plt.xlabel('Season')
plt.xticks(rotation=90)
plt.ylabel('flower types')
plt.show()
```

```python
pie = data["Features"].value_counts()
pie.plot(kind = "pie", autopct = "%1.1f%%")
plt.show()
```
