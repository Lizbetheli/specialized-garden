```python
import pandas as pd
```

```python
df=pd.read_csv('../data/machine_learning.csv', index_col=0)
df.head()
# df=df.applymap(lambda x: x.strip())
```

```python
text_data=df.select_dtypes(object).columns
# df.select_dtypes(object).applymap(lambda x: x.strip() if type(x)==str else x)
df[text_data]=df[text_data].applymap(lambda x: x.strip() if type(x)==str else x)
df[text_data]=df[text_data].applymap(lambda x: x.replace('\u2009', ' ') if type(x)==str else x)
df[text_data]=df[text_data].applymap(lambda x: x.replace('\xa0', ' ') if type(x)==str else x)
```

```python
df.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder()
# oh.fit(df['Plant Type'].values)
encoded=oh.fit_transform(df['Plant Type'].values.reshape(-1, 1))
plant_type_df=pd.DataFrame(encoded.toarray(), columns=oh.get_feature_names(['Plant Type']))
plant_type_df
```

```python
oh=OneHotEncoder()
encoded=oh.fit_transform(df['Season'].fillna('All').values.reshape(-1, 1))
season_df=pd.DataFrame(encoded.toarray(), columns=oh.get_feature_names(['Season']))
season_df
```

```python
states_df=df[['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', \
              'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', \
              'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', \
              'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', \
              'New Jersey', 'New Mexico', 'New york', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', \
              'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', \
              'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming']].fillna(0)
states_df
```

```python
oh=OneHotEncoder()
encoded=oh.fit_transform(df[['Sun Needs', 'Soil Needs']].fillna('Flexible').values.reshape(-1, 2))
needs_df=pd.DataFrame(encoded.toarray(), columns=oh.get_feature_names(['Sun Needs', 'Soil Needs']))
needs_df
```

```python
# from sklearn.preprocessing import LabelEncoder
# lb=LabelEncoder()
# lb.fit(df['Plant Type'])
# plant_type_df=pd.DataFrame(lb.transform(df['Plant Type']), columns=['Plant Type'])
# plant_type_df
```

```python
features_df=df[['Tolerates heat', 'Tolerates frost',
       'Non-invasive', 'Good for pressed flowers', 'Good for cut flowers',
       'Deer resistant', 'Attracts bees', 'Attracts butterflies',
       'Attracts beneficial insects', 'Attracts hummingbirds']].fillna(0)
features_df
```

```python
# input_df=pd.concat([states_df, plant_type_df], axis=1)
# input_df=pd.concat([states_df], axis=1)
# input_df=pd.concat([states_df, df['Plant Type']], axis=1)
input_df=pd.concat([states_df, plant_type_df, season_df, needs_df, features_df], axis=1)
input_df
```

```python
input_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
output_df=df['Plant Name']
```

```python
input_df.to_csv('../data/ml_input.csv')
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(input_df, output_df)
clf.score(input_df, output_df)
```

```python
clf.classes_
```

```python
prediction=clf.predict(input_df)
results=pd.DataFrame({'Prediction': prediction, 'Actual': output_df})
results['Accurate']=results['Prediction']==results['Actual']
results[results['Accurate']!=True]
```
